## Step 3: Feature Engineering⚙️

Feature engineering involves creating new meaningful features from existing data to improve analysis and model performance.

In this step, we create:

* Total Marks
* Average Marks
* Grade classification
* Pass/Fail status


```python
df['Total_Marks'] = df['Math_Score'] + df['Science_Score'] + df['English_Score']  # Creating Total Marks column
df.tail()  # Display updated data
```

```python
df['Average_Marks']=df['Total_Marks']/3
```

### 🎓 Grade Classification

Grades are assigned based on average marks to categorize student performance.


```python
def assign_grade(avg):   #Function to assign grade
    if avg >= 90:
        return 'A'
    elif avg >= 75:
        return 'B'
    elif avg >= 60:
        return 'C'
    else:
        return 'D'

df['Grade']=df['Average_Marks'].apply(assign_grade)  #Apply Function
```

```python
df['Result']=df['Average_Marks'].apply(lambda x: 'Pass' if x>= 40 else 'Fail')  #Pass/Fail Classification
```

```python
df.tail(56)
```

### 🔎 Feature Engineering Insights

* Total and average marks provide a better understanding of overall performance
* Grade classification simplifies performance comparison
* Pass/Fail status helps identify students needing improvement

These features enhance the dataset and make it more suitable for analysis and modeling.

## Step 7: Machine Learning 🤖

In this step, a Linear Regression model is built to predict student total marks based on study hours and attendance.

This demonstrates how Machine Learning can be used to analyze and predict academic performance.


```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split  #Splits data
from sklearn.linear_model import LinearRegression     # ML Model
from sklearn.metrics import mean_absolute_error, r2_score  # Measure prediction error & Measure model accuracy
```

```python
X = df[['Study_Hours', 'Attendance']]   #Factors used for prediction(inputs)
Y = df['Total_Marks']                   #what we want to predict(output)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(        # Splitting dataset into training and testing data
    X,
    Y,
    test_size=0.2,
    random_state=42
)
```

```python
model = LinearRegression()      # Creating Linear Regression model
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)    # Predicting total marks
```

```python
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print("Mean Absolute Error:", mae)
print("R2 Score:", r2)
```

```python
results = pd.DataFrame({
    'Actual Marks':Y_test,
    'Predicted Marks':Y_pred
})
results.head()
```

## 📊 Machine Learning Insights

* The Linear Regression model successfully predicted student total marks using study hours and attendance.
* Students with higher attendance and study hours generally achieved better academic performance.
* The model demonstrates how Machine Learning can be applied in educational analytics for performance prediction.
